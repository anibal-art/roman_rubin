```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Laplacian function
def laplacian(theta1, theta2, theta_c, epsilon):
    numerator1 = (1 - epsilon) * (theta_c**2 + (1 + epsilon) * theta2**2)
    numerator2 = (1 + epsilon) * (theta_c**2 + (1 - epsilon) * theta1**2)
    denominator = (theta_c**2 + (1 - epsilon) * theta1**2 + (1 + epsilon) * theta2**2)**(3/2)
    return (numerator1 + numerator2) / denominator

# Define parameters
theta_c = 1.0  # Core radius
epsilon = 0.5  # Ellipticity
theta1_range = np.linspace(-2, 2, 200)
theta2_range = np.linspace(-2, 2, 200)

# Create a meshgrid for theta1 and theta2
theta1, theta2 = np.meshgrid(theta1_range, theta2_range)

# Calculate the Laplacian
laplacian_values = laplacian(theta1, theta2, theta_c, epsilon)

# Plot the Laplacian
plt.figure(figsize=(8, 6))
plt.contourf(theta1, theta2, laplacian_values, levels=5, cmap='viridis')
plt.colorbar(label='Laplacian')
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'$\theta_2$')
plt.title(f'Laplacian of $\Psi$ with $\\theta_c = {theta_c}$, $\\epsilon = {epsilon}$')
plt.grid(alpha=0.3)
plt.show()
```

```python

def ellipse(theta1, theta2, epsilon):
    return (theta1**2)/(1 + epsilon)**(2) +  (theta2**2)/(1-epsilon)**(2)

# Calculate the radial values for the ellipses
ellipse_values = ellipse(theta1, theta2, epsilon)

# Plot the Laplacian with ellipse contours
plt.figure(figsize=(8, 6))
contour = plt.contourf(theta1, theta2, laplacian_values, levels=6, cmap='viridis')
plt.colorbar(contour, label='Laplacian')
plt.contour(theta1, theta2, ellipse_values, levels=[1, 2, 3, 4], colors='red', linewidths=1.5)
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'$\theta_2$')
plt.title(f'Laplacian of $\Psi$ with Ellipse Contours ($\\theta_c = {theta_c}$, $\\epsilon = {epsilon}$)')
plt.grid(alpha=0.3)
plt.savefig('/home/anibal-pc/elipticidad_'+str(epsilon)+'.png')
plt.show()
```
