```python
import numpy as np
import matplotlib.pyplot as plt

data_fit_rr = np.load('/home/anibal/roman_rubin/MCMC_fit/Event_RR_18_MCMC.npy', allow_pickle=True).item()

print(data_fit_rr['true_params'])
plt.plot(data_fit_rr['MCMC_chains'][:,:,8])
plt.show()
```
